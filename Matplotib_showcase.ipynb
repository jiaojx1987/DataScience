```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)
```

```python
# line plot
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Scatter Plot
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Bar plot
categories = ["A", "B", "C", "D", "E"]
values = [15,7,10,12,8]
plt.bar(categories, values)
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Histogram
plt.hist(y,bins=20)
plt.title("Histogarm")
plt.xlabel('y range')
plt.ylabel('Frequency')
plt.show()
```

```python
explode = [0,0,.1,0,0]
plt.pie(values,explode=explode,labels=categories)
plt.title("Pie Chart")
plt.show()
```

```python
plt.fill_between(x,y,alpha = .5)
plt.title("Area Plot")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.show()
```

```python
ohic = np.vstack((x,y,y,y))
# combine plot
plt.plot(x,y,color='b')
plt.stem(x,y,'r',markerfmt='ro')
plt.title("Candlestick Chart")
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.show()
```

```python
plt.violinplot(y,showmedians = True)
plt.title("violin Plot")
plt.ylabel("Y_axis")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d")
ax.plot_trisurf(x,y,x**2+y**2, cmap='viridis')
ax.set_title("3D Surface Plot")
ax.set_xlabel("X_axis")
ax.set_ylabel("Y_axis")
ax.set_zlabel("Z-axis")
plt.show()
```

```python
plt.hexbin(x,y,gridsize=20,cmap='inferno')
plt.title("Hexbin Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(label='Count')
plt.show()
```

```python
theta = np.linspace(0,2*np.pi, 100)
r = 2 + np.sin(5*theta)
```

```python
plt.polar(theta, r)
plt.title("Polar Plot")
plt.show()
```

```python
x_1, y_1 = np.meshgrid(np.linspace(-3,3,100), np.linspace(-3,3,100))
u = -y_1
v = x_1
```

```python
plt.streamplot(x_1,y_1,u,v)
plt.title("Streamplot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
errors = abs(np.random.normal(0,.2,100))
```

```python
plt.errorbar(x,y,yerr=errors, fmt="o",capsize=3, ecolor="r")
plt.title("Errorbar Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
